```python
# 강우로 인한 미세먼지 농도 차이가 있을 것이다

import csv
import matplotlib.pyplot as plt

f = open('daegu_pm.csv')
data = csv.reader(f)
header = next(data)
#print(header)

pm_10 = []
pm_25 = []
rain = []


for row in data:
    pm_10.append(int(row[1]))
    pm_25.append(int(row[2]))
    rain.append(float(row[3]))

#print(pm_10)
#print(pm_25)

plt.figure(figsize = (15, 5))
plt.rc("font", family = "Malgun Gothic")
plt.title("대구광역시 2020년 1월 - 3월 미세먼지와 강우량", pad = 12)
plt.plot(pm_10, label = "PM10" , color = "orange")
plt.plot(pm_25, label= "PM2.5", color = "gold")
plt.bar(range(91), rain, label = "Rain(mm)", color = "blue", width = 1.2)
plt.legend()
plt.show()
```
