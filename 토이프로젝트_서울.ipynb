```python
# 계절별미세먼지 농도 추이
## 봄철 미세먼지 농도가 가장 높을 것이다
import csv
import matplotlib.pyplot as plt

f = open('seoul_pm.csv')
data = csv.reader(f, delimiter = ',')
header = next(data)
#print(header)

s_pm10_w = [] # 겨울
s_pm10_sp = [] # 봄
s_pm10_sm = [] # 여름
s_pm10_f = [] # 가을

for row in data:
    if "평균" in row[3]:
        if row[1].split('-')[0] == "2020":
            if row[1].split('-')[2] == "01" or row[1].split('-')[2] == "02" or row[1].split('-')[2] == "12" :
                s_pm10_w.append(int(row[4]))
            elif row[1].split('-')[2] == "03" or row[1].split('-')[2] == "04" or row[1].split('-')[2] == "05" :
                s_pm10_sp.append(int(row[4]))
            elif row[1].split('-')[2] == "06" or row[1].split('-')[2] == "07" or row[1].split('-')[2] == "08" :
                s_pm10_sm.append(int(row[4]))
            elif row[1].split('-')[2] == "09" or row[1].split('-')[2] == "10" or row[1].split('-')[2] == "11" :
                s_pm10_f.append(int(row[4]))
#print(len(s_pm10_w))
#print(len(s_pm10_sp))
#print(len(s_pm10_sm))
#print(len(s_pm10_f))
        
#그래프그리기
plt.figure(figsize = (15, 4))
plt.rc("font", family = "Malgun Gothic")
plt.title("계절별 미세먼지 농도 추이")
plt.plot(s_pm10_sp, label = "봄", color = "gold")
plt.plot(s_pm10_sm, label = "여름", color = "limegreen")
plt.plot(s_pm10_f, label = "가을", color = "orange")
plt.plot(s_pm10_w, label = "겨울", color = "skyblue")
plt.legend()
plt.show()
```

```python
# 계절별 -2 (2021년이랑 섞어서)
import csv
import matplotlib.pyplot as plt

f = open('seoul_pm.csv')
data = csv.reader(f, delimiter = ',')
header = next(data)
#print(header)

s_pm10_w = []
s_pm10_sp = []
s_pm10_sm = []
s_pm10_f = []

for row in data:
    if "평균" in row[3]:
        if row[1].split('-')[0] == "2020":
            if row[1].split('-')[2] == "12":
                s_pm10_w.append(int(row[4]))
            elif row[1].split('-')[2] == "03" or row[1].split('-')[2] == "04" or row[1].split('-')[2] == "05" :
                s_pm10_sp.append(int(row[4]))
            elif row[1].split('-')[2] == "06" or row[1].split('-')[2] == "07" or row[1].split('-')[2] == "08" :
                s_pm10_sm.append(int(row[4]))
            elif row[1].split('-')[2] == "09" or row[1].split('-')[2] == "10" or row[1].split('-')[2] == "11" :
                s_pm10_f.append(int(row[4]))
        if row[1].split('-')[0] == "2021" :
            if row[1].split('-')[2] == "01" or row[1].split('-')[2] == "02":
                s_pm10_w.append(int(row[4]))
            
# 그래프 그리기
plt.figure(figsize = (15, 5))
plt.plot(s_pm10_w, label = "winter")
plt.plot(s_pm10_sp, label = "spring")
plt.plot(s_pm10_sm, label = "summer")
plt.plot(s_pm10_f, label = "fall")
plt.legend()
plt.show()
```
